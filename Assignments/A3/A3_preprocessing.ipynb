```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import json
```

```python
filedata = []
with open("data/video_games_700k.jsonl") as f:
    for line in f:
        json_object = json.loads(line)
        filedata.append(json_object)

data = pd.DataFrame(filedata)
```

```python
data
```

```python
data.rating.value_counts().plot(kind='bar')
```

```python
data.rating.value_counts().plot(kind='pie')
```

```python
data.rating.value_counts()
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
data.columns
```

```python
data.drop(columns=['images', 'user_id'], inplace=True)
```

```python
data['text'] = data['title']+ ' ' + data['text']
```

```python
data.drop(columns=['title'], inplace=True)
```

```python
data
```

```python
data = data[data['verified_purchase']]
```

```python
data
```

```python
data.rating.value_counts()
```

```python
data = data[data['rating'] != 3.0]
```

```python
data.rating.value_counts()
```

```python
data['sentiment'] = data['rating'].apply(lambda x: 'positive' if x > 3.0 else 'negative')
```

```python
data.sentiment.value_counts().plot(kind='bar')
```

```python
data.sentiment.value_counts()
```

```python
negative_sentiments = data[data['sentiment'] == 'negative']
positive_sentiments = data[data['sentiment'] == 'positive']

positive_sample = positive_sentiments.sample(n=len(negative_sentiments), random_state=42)
```

```python
data = pd.concat([positive_sample, negative_sentiments])
```

```python
data
```

```python
data.text = data.text.str.lower()
```

```python
data
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))

data.text = data.text.apply(lambda x: ' '.join([word for word in word_tokenize(x) if word not in stop_words]))
```

```python
data
```

```python
data.drop(columns=['verified_purchase', 'rating'], inplace=True)
```

```python
data
```

```python
data['text'] = data['text'].str.replace('[^\w\s]', '')
```

```python
data
```

```python
data.to_json('data/video_games_700k_cleaned.jsonl', orient='records', lines=True)
```

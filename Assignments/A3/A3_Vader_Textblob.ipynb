```python
# Step 3

# --------------------------------------------------------------------
# Run TextBlob on the reviews
# --------------------------------------------------------------------
import pandas as pd
import nltk
nltk.download('vader_lexicon')
from textblob import TextBlob
from nltk.sentiment.vader import SentimentIntensityAnalyzer


dataset = pd.read_csv('/Users/aishwaryaravichandran/Documents/nlp/nlp_assignment_0_group_work/CS6120/Assignments/A3/review_set.csv')
print(dataset.shape)
df = dataset.dropna(subset=['review'], inplace=True)
dataset.reset_index(drop=True, inplace=True)
print(dataset.shape)

analyzer = SentimentIntensityAnalyzer()

# textblob
def textblob_polarity(text):
    return TextBlob(text).sentiment.polarity

dataset['textblob_polarity'] = dataset['review'].apply(textblob_polarity)

# Adding textblob subjectivity
def textblob_subjectivity(text):
    return TextBlob(text).sentiment.subjectivity

dataset['textblob_subjectivity'] = dataset['review'].apply(textblob_subjectivity)

dataset.head()
```

```python
# --------------------------------------------------------------------
# Run Vader on the reviews
# --------------------------------------------------------------------

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer


analyzer = SentimentIntensityAnalyzer()

def vader_polarity(text):
    return analyzer.polarity_scores(text)['compound']
# Calculate the breakdown of the sentiment
def vader_positive(text):
    return analyzer.polarity_scores(text)['pos']
def vader_neutral(text):
    return analyzer.polarity_scores(text)['neu']
def vader_negative(text):
    return analyzer.polarity_scores(text)['neg']

dataset['vader_positive'] = dataset['review'].apply(vader_positive)
dataset['vader_neutral'] = dataset['review'].apply(vader_neutral)
dataset['vader_negative'] = dataset['review'].apply(vader_negative)

dataset['vader_polarity'] = dataset['review'].apply(vader_polarity)

dataset.head()
```

```python
dataset['vader_sentiment'] = dataset['vader_polarity'].apply(lambda x: 'positive' if x > 0.5 else 'negative')
dataset['textblob_sentiment'] = dataset['textblob_polarity'].apply(lambda x: 'positive' if x > 0 else 'negative')

dataset.head()
```

```python
dataset.shape
```

```python
# --------------------------------------------------------------------
# Predicting Vader Accuracy
# --------------------------------------------------------------------

total_predictions = 209925
vader_predictions = (dataset['vader_sentiment'] == dataset['sentiment']).sum()
vader_accuracy = vader_predictions / total_predictions
print(vader_accuracy)
```

```python
# --------------------------------------------------------------------
# Predicting Textblob Accuracy
# --------------------------------------------------------------------

textBlob_predictions = (dataset['textblob_sentiment'] == dataset['sentiment']).sum()
textblob_accuracy = textBlob_predictions / total_predictions
print(textblob_accuracy)
```

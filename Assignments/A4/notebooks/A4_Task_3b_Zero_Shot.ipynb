```python
def categorize_text_column(df: pd.DataFrame) -> pd.DataFrame:
    # Load the zero-shot classification pipeline
    categories = ['restaurants', 'bars', 'coffee shops',
                  'hotels', 'salons', 'auto repair', 'home services', 'medical services',
                  'entertainment', 'pet services', 'financial services',
                  'travel and tourism', 'education', 'real estate', 'fitness',
                  'landscape and gardending', 'legal services', 'photography services',
                  'computer and technology services']

    for row in range(len(df)):
        predicted_category = classifier(df['text'][row], candidate_labels=categories, multi_label=True)['labels'][0]
        print(predicted_category)
        df.loc[row, predicted_category] = 1
        
    return df

zero_shot = categorize_text_column(zero_shot)
```

```python
zero_shot.drop(columns=['label'], inplace=True)
manual_data.drop(columns=['label'], inplace=True)
```

```python
count = 0
for row in range(len(zero_shot)):
    for column in zero_shot.columns:
        if zero_shot[column][row] == manual_data[column][row] == 1:
            count = count + 1
print(count)
```
